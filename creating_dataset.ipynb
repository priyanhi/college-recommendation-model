```python
# HERE WE START OUR DATA PREPROSSAING/ DATA TRANSFORMATION/ DATA RESHAPING

# DATA RESHAPING MEAN:-- convert raw data into a format that is easier for analysis or ML 

# LOAD CSV
# LOAD CSV
eapcte_df = pd.read_csv("final_third_phase.csv")

# -----------------------------
# SET FIRST ROW AS HEADER
# -----------------------------
eapcte_df.columns = eapcte_df.iloc[0]

# REMOVE FIRST ROW
eapcte_df = eapcte_df[1:].reset_index(drop=True)

# -----------------------------
# CLEAN COLUMN NAMES
# -----------------------------
eapcte_df.columns = (
    eapcte_df.columns
    .astype(str)
    .str.replace("\n", " ", regex=False)
    .str.replace(r"\s+", " ", regex=True)
    .str.strip()
)

# -----------------------------
# REMOVE REPEATED HEADER ROWS
# -----------------------------
eapcte_df = eapcte_df[
    ~eapcte_df["Inst Code"]
    .astype(str)
    .str.contains("Inst", case=False, na=False)
]

# -----------------------------
# REMOVE EMPTY ROWS
# -----------------------------
eapcte_df = eapcte_df.dropna(how="all")

# -----------------------------
# RESET INDEX
# -----------------------------
eapcte_df = eapcte_df.reset_index(drop=True)

# -----------------------------
# CHECK CLEANED COLUMNS
# -----------------------------
print(eapcte_df.columns.tolist())

# -----------------------------
# RANK COLUMNS
# -----------------------------
rank_columns = [
    "OC BOYS", "OC GIRLS",

    "BC_A BOYS", "BC_A GIRLS",
    "BC_B BOYS", "BC_B GIRLS",
    "BC_C BOYS", "BC_C GIRLS",
    "BC_D BOYS", "BC_D GIRLS",
    "BC_E BOYS", "BC_E GIRLS",

    "SC_I BOYS", "SC_I GIRLS",
    "SC_II BOYS", "SC_II GIRLS",
    "SC_III BOYS", "SC_III GIRLS",

    "ST BOYS", "ST GIRLS",

    "EWS BOYS", "EWS GIRLS"
]

# -----------------------------
# MELT DATA
# -----------------------------
long_df = eapcte_df.melt(
    id_vars=[
        "Inst Code",
        "Institute Name",
        "Place",
        "Dist Code",
        "Co Education",
        "College Type",
        "Branch Code",
        "Branch Name",
        "Affiliated To"
    ],

    value_vars=rank_columns,

    var_name="Category_Gender",

    value_name="Closing_Rank"
)

# -----------------------------
# SPLIT CATEGORY AND GENDER
# -----------------------------
long_df[["Category", "Gender"]] = (
    long_df["Category_Gender"]
    .str.rsplit(" ", n=1, expand=True)
)

# -----------------------------
# DROP EXTRA COLUMN
# -----------------------------
long_df.drop(columns=["Category_Gender"], inplace=True)

# -----------------------------
# REMOVE NULL VALUES
# -----------------------------
long_df = long_df.dropna(subset=["Closing_Rank"])

# REMOVE EMPTY STRINGS
long_df = long_df[
    long_df["Closing_Rank"]
    .astype(str)
    .str.strip() != ""
]
```

```python
#SECOND

#Some rows have:

#Branch Code = NaN

#like:

#JBIT NaN
#MGIT NaN

#These are also broken extraction rows.

#Remove them:


long_df = long_df[
    long_df["Branch Code"]
    .astype(str)
    .str.strip() != ""
]
```

```python
#THIRD FIX — Clean ALL important columns

#Before ANY merge:

important_cols = [
    "Inst Code",
    "Branch Code"
]

for col in important_cols:

    long_df[col] = (
        long_df[col]
        .astype(str)
        .str.strip()
        .str.upper()
    )
```

```python
# make sure numeric
long_df["Closing_Rank"] = pd.to_numeric(long_df["Closing_Rank"], errors="coerce")
long_df["Seats"] = pd.to_numeric(long_df["Seats"], errors="coerce")

# competition ratio
long_df["Competition_Ratio"] = (
    long_df["Closing_Rank"] / long_df["Seats"]
)
```

```python
#Branch Demand Score

branch_demand = long_df.groupby("Branch Code")["Closing_Rank"].mean()

long_df["Branch_Demand"] = long_df["Branch Code"].map(branch_demand)
```

```python

#Institute Difficulty Score

inst_difficulty = long_df.groupby("Inst Code")["Closing_Rank"].mean()

long_df["Institute_Difficulty"] = long_df["Inst Code"].map(inst_difficulty)
```

```python

#Category Competition Pressure

category_pressure = long_df.groupby("Category")["Closing_Rank"].mean()

long_df["Category_Pressure"] = long_df["Category"].map(category_pressure)
```

```python

#Rank Normalization


from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

long_df["Normalized_Closing_Rank"] = scaler.fit_transform(
    long_df[["Closing_Rank"]]
)
```
