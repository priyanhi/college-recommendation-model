```python
df = pd.read_csv("EAPCET_dataset.csv")
```

```python
# create a  Recommendation score

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler() # this is object from minmaxscaler class
df[[ # double brackit are used because multiple columns are involved
    
    "Branch_Demand_N",
    "Institute_Difficulty_N",
    "Competitive_Percentile_N" # selected multiple columns in a Pandas Data Frame
]] = scaler.fit_transform(df[[
    "Branch_Demand",
    "Institute_Difficulty",
    "Competitive_Percentile"
]])

df["Recommendation_Score"] = (
      0.5 * df["Branch_Demand_N"]
    + 0.3 * df["Institute_Difficulty_N"]
    + 0.2 * df["Competitive_Percentile_N"]
)
```

```python
# select features

features = [
    "Institute Name",
    "Place",
    "College Type",
    "Branch Name",
    "Category",
    "Gender",
    "ROUND",
    "year",
    "Closing_Rank",
    "Prev_Year_Closing_Rank",
    "Opening_Rank",
    "Branch_Demand",
    "Institute_Difficulty",
    "Category_Pressure",
    "Normalized_Closing_Rank",
    "Competitive_Percentile",
    "NIRF_Rank",
    "Institute_Tier"
]

x = df[features]

y = df["Recommendation_Score"]
```

```python
# separate Numerical and Categorical Features
categorical_features  = [
    "Institute Name",
    "Place",
    "College Type",
    "Branch Name",
    "Category",
    "Gender",
    "NIRF_Rank",
    "Institute_Tier"
]

numerical_features = [
    "ROUND",
    "year",
    "Closing_Rank",
    "Prev_Year_Closing_Rank",
    "Opening_Rank",
    "Branch_Demand",
    "Institute_Difficulty",
    "Category_Pressure",
    "Normalized_Closing_Rank",
    "Competitive_Percentile"
]
```

```python
# preprocessing

preprossing = ColumnTransformer(
    transformers =[
        (
            "cat",
            OneHotEncoder(handle_unknown = "ignore"),
            categorical_features
        )
    ],
    remainder = "passthrough"
)
```

```python
# Train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size = 0.2,
    random_state = 42
)
```

```python
# building pipeline

pipeline = Pipeline([
    ("preprocessing", preprossing),
    (
        "model",
        RandomForestRegressor(
            n_estimators=300,
            max_depth=15,
            random_state=42,
            n_jobs=-1
        )
    )
])
```

```python
# tran_model

pipeline.fit(x_train, y_train)
```

```python
# Get Recommendations


user_rank = 15000
user_category = "OC"
user_gender = "BOYS"
user_branch = "COMPUTER SCIENCE AND ENGINEERING"
```

```python
# predict probability

eligible = df[
    (df["Category"] == user_category)
    &
    (df["Gender"] == user_gender)
    &
    (df["Branch Name"] == user_branch)
    &
    (df["Closing_Rank"] >= user_rank)
].copy()
```

```python
# predict 

eligible["Predicted_Score"] = pipeline.predict(
    eligible[features]
)
```

```python
def recommend_colleges(rank, category, gender, branch):

    # Filter eligible colleges
    eligible = df[
        (df["Category"] == category)
        &
        (df["Gender"] == gender)
        &
        (df["Branch Name"] == branch)
        &
        (df["Closing_Rank"] >= rank)
    ].copy()

    # Check if colleges found
    if len(eligible) == 0:
        print("No colleges found for given criteria")
        return

    # Predict recommendation score
    eligible["Predicted_Score"] = pipeline.predict(
        eligible[features]
    )

    # Calculate rank difference
    eligible["Rank_Difference"] = (
        eligible["Closing_Rank"] - rank
    )

    # Remove duplicates and keep best score
    final_output = (
        eligible
        .groupby(
            [
                "Institute Name",
                "Branch Name",
                "Place"
            ],
            as_index=False
        )
        .agg({
            "Closing_Rank": "max",
            "Predicted_Score": "max",
            "Rank_Difference": "min"
        })
        .sort_values(
            "Predicted_Score",
            ascending=False
        )
    )

    # Display top 10
    return final_output[
        [
            "Institute Name",
            "Branch Name",
            "Place",
            "Closing_Rank",
            "Rank_Difference",
            "Predicted_Score"
        ]
    ].head(10)
```

```python



from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor

from sklearn.ensemble import(
    RandomForestRegressor,
    GradientBoostingRegressor
)
```

```python
models = {
    
    "Random Forest": RandomForestRegressor(
        n_estimators = 300,
        random_state = 42
    ),

    "Gradient Boosting": GradientBoostingRegressor(
        random_state = 42
    ),

    "XGBoost": XGBRegressor(
        random_state = 42,
        verbose = 0
    ),

    "CatBoot": CatBoostRegressor(
        random_state = 42,
        verbose = 0
    ),

    "LightGBM": LGBMRegressor(
        random_state = 42
    )
}
```

```python
from sklearn.metrics import (
    r2_score,
    mean_absolute_error,
    mean_squared_error
)

results = []

for name, model in models.items():
    
    pipe = Pipeline([
        ("preprossing", preprossing),
        ("model", model)
    ])

    pipe.fit(x_train, y_train)

    predictions_1 = pipe.predict(x_test)
    results.append({
        "Model": name,

        "R2 Score": r2_score(
            y_test,
            predictions_1
        ),

        "MAE": mean_absolute_error(
            y_test,
            predictions_1
        ),

        "MSE": mean_squared_error(
            y_test,
            predictions_1
        )
    })

    results = pd.DataFrame(results)

    results.sort_values(
        "R2_score",
        ascending = False
    )
```
